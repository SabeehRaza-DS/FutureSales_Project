```python
# EDA 

def EDA(df):
    print(" ")
    print("----------TOP 5 RECORDS--------")
    print(df.head(5))
    print(" ")
    print("----------Data Types-----------")
    print(df.dtypes)
    print(" ")
    print("----------INFO-----------------")
    print(df.info())
    print(" ")
    print("-----Data Shape-------------")
    print(df.shape)
    print(" ")
    print("-------Missing Values----------")
    print(df.isnull().sum())
    print(" ")
    print("-------NULL values-------------")
    print(df.isna().sum())
    
    
```

```python

print("*************** Item Categories ***************")
EDA(df_itm_cat)
```

```python
print("*************** Items ***************")
EDA(df_itm)
```

```python
print("*************** Shops ***************")
EDA(df_sh)
```

```python
print("*************** Training Sales Data ***************")
EDA(df_tr)
```

```python
print("*************** Test data ***************")
EDA(df_te)
```
